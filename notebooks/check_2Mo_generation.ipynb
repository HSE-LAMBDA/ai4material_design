```python
from pathlib import Path
from pymatgen.io.vasp.inputs import Poscar
```

```python
path = Path("../datasets/POSCARs/MoS2-2Mo/1V_1W_Mo/poscars")
```

```python
# load every file in path as POSCAR
structures = [Poscar.from_file(file).structure for file in path.iterdir()]
```

```python
for i in range(len(structures)):
    for j in range(i+1, len(structures)):
        if structures[i].matches(structures[j]):
            print(f"Found matching structures {i} and {j}")
```
