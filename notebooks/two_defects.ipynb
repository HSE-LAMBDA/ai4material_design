```python
import pandas as pd
```

```python
predicted = pd.read_csv(
    "ai4material_design/datasets/predictions/MoS2_train_4_5_6_test_8/band_gap/megnet_pytorch-sparse-z-were.csv.gz",
    index_col="_id"
)
predicted.head()
```

```python
predicted_cv = pd.read_csv(
    "ai4material_design/datasets/predictions/MoS2-plain-cv/band_gap/megnet_pytorch-sparse-z-were.csv.gz",
    index_col="_id"
)
predicted_cv.head()
```

```python
targets = pd.read_csv(
    "ai4material_design/datasets/processed/dichalcogenides_MoS2_8x8/targets.csv.gz",
    index_col="_id"
)
targets
```

```python
matches = pd.read_csv("ai4material_design/datasets/csv_cif/dichalcogenides_MoS2_8x8/defects.csv")
matches
```

```python
cur_structs = matches[matches["descriptor_id"] == "6141cf171d648932fbc3428d"]["_id"].tolist()
cur_structs
```

```python
data = pd.read_pickle("ai4material_design/datasets/processed/dichalcogenides_MoS2_8x8/data.pickle.gz")
data["defect_representation"]
```

```python
structures = data[data.index.isin(cur_structs)]["defect_representation"].copy()
```

```python
structures = structures.apply(lambda x: x.get_distance(0, 1))
```

```python
structures
```

```python
cur_targets = targets[targets.index.isin(cur_structs)]
cur_predictions = predicted[predicted.index.isin(cur_structs)]
cur_predictions_cv = predicted_cv[predicted_cv.index.isin(cur_structs)]
```

```python
target_line = pd.merge(structures, 
                          cur_targets, 
                          left_index=True, 
                          right_index=True)[['defect_representation', 'band_gap']].set_index("defect_representation")
```

```python
predicted_line = pd.merge(structures, 
                          cur_predictions, 
                          left_index=True, 
                          right_index=True)[['defect_representation', 'predicted_band_gap_test']].set_index("defect_representation")
```

```python
predicted_line_cv = pd.merge(structures, 
                          cur_predictions_cv, 
                          left_index=True, 
                          right_index=True)[['defect_representation', 'predicted_band_gap_test']].set_index("defect_representation")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(10, 5))

plt.plot(target_line.sort_index(), label="real", linestyle='--', marker='o')
plt.plot(predicted_line.sort_index(), label="train on 4, 5, 6", linestyle='--', marker='o')
plt.plot(predicted_line_cv.sort_index(), label="crossvalidate", linestyle='--', marker='o')

plt.grid()

plt.title("two substitutions Mo->W S->Se")
plt.xlabel("distance between defects")
plt.ylabel("band gap")
plt.xticks(np.arange(3, 15))
plt.legend();
```
