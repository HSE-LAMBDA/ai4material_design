```python
import ase.io
from ase.visualize.plot import plot_atoms
import matplotlib.pyplot as plt
```

```python
ase_atoms = ase.io.read("hBN.cif")
fig, axes = plt.subplots(1, 2, figsize=(10, 3.5))
plot_atoms(ase_atoms, ax=axes[0], radii=0.7, rotation=('0x, 0y, 0z'), show_unit_cell=False)
plot_atoms(ase_atoms, ax=axes[1], radii=0.7, rotation=('90x, 0y, 0z'), show_unit_cell=False)
for ax in axes:
    ax.set_xlabel('$\AA$')
    ax.set_ylabel('$\AA$')
str_description = []
axes[0].set_title("Top view")
axes[1].set_title("Side view")
title = f"hBN with 7 B vacancies"
fig.suptitle(title)
fig.savefig("hBN.pdf", bbox_inches='tight')
fig.savefig("hBN.png", bbox_inches='tight', dpi=300)
```
