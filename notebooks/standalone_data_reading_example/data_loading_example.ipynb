# Load all the structures
A quick way is to use the non-future proof pickle format. The files have been prepared by running the Rolos workflow available in this project.

```python
structures_pkl = pd.read_pickle("structures.pkl.gz")
defects_pkl = pd.read_pickle("defects.pkl.gz")
print(f"Pickle has {len(structures_pkl)} structures.")
```

A more robust and long way is reading all the cif files anew:

```python
datasets = [f"datasets/csv_cif/high_density_defects/{name}_500" for name in ("BP_spin", "GaSe_spin", "hBN_spin", "InSe_spin", "MoS2", "WSe2")] + \
           [f"datasets/csv_cif/low_density_defects/{name}" for name in ("MoS2", "WSe2")]
```

```python
from multiprocessing import Pool
# Shame tqdm.notebook has issues with vscode
# https://github.com/microsoft/vscode-jupyter/issues/11014
with Pool(processes=8) as pool:
    data = pool.map(read_csv_cif, datasets)
```

```python
structures, defects = map(pd.concat, zip(*data))
```

```python
print(f"We have {len(structures)} structures in csv/cif and {len(structures_pkl)} in pickle")
```
